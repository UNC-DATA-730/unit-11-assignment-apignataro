Let's build a random forest model to predict `MedianHouseValue`.

1. Create your model, set the engine to `ranger` with `importance = 'impurity'` and the mode as `regression`.
2. Fit the model using `MedianHouseValue` as the response variable and all other columns as predictors.

````r
## This will split my model into training and test sets, using set.seed(123) for reproducibility
set.seed(123)
split <- initial_split(housing, prop = 0.8)
train_data <- training(split)
test_data <- testing(split)

## This will buil the random forest model setting the engine to ranger w/ importance = 'impurity'
## mode as regression
rf_model <- rand_forest(mode = "regression") %>%
  set_engine("ranger", importance = "impurity")

## This will fit our rf model
rf_fit <- rf_model %>%
  fit(MedianHouseValue ~ ., data = train_data)
````

❓ What is the OOB error for this model?

**Answer:**

````r
## This will extract the ranger engine
rf_fit_model <- rf_fit %>% 
  extract_fit_engine()

## This will output the OOB error for this model
rf_fit_model$prediction.error
````

Plot the feature importances using the `vip` package. Remember that you can use the `extract_fit_engine` function to get the `ranger` model object to send to the `vip` function.

````r
library('vip')

## This will extract the ranger engine
rf_fit_model <- rf_fit %>%
  extract_fit_engine()

## This will plot the feature importances using the vip package
vip(rf_fit_model)
````

````r
## This will load the necessary libraries
library(ggplot2)
library(dplyr)

california <- map_data("state") |> filter(region == "california")

## This will plot a map of California w/ Latitude and Longitude from the housing dataset
## This will color points by the MedianHouseValue
## used the magma color scale to have the map be similar to the example
ggplot(data = california) +
  geom_polygon(aes(x = long, y = lat), fill = "grey90", color = "grey40") +
  geom_point(data = housing, 
             aes(x = Longitude, y = Latitude, color = MedianHouseValue), 
             alpha = 0.6, size = 1) +
  coord_quickmap() +
  scale_color_viridis_c(option = "magma") +  
  theme_void() +
  labs(color = "MedianHouseValue",
       title = "California Housing Prices Across the State")
````
